```python

def sum_tuples(t1, t2):
    return tuple(a+b for a, b in zip(t1, t2))

def sub_tuples(t1, t2):
    return tuple(a-b for a, b in zip(t1, t2))
```

```python
data = day1.read_data_string("input.txt")
# data = day1.read_data_string("input_sample.txt")
lines = day1.split_data(data) 
paragraphs = [day1.split_data_p(paragraph) for paragraph in day1.split_data_paragraphs(data)]
```

```python
def parse_line(line): 
    b1, b2 = line.split("~")
    b1 = tuple(int(b) for b in b1.split(","))
    b2 = tuple(int(b) for b in b2.split(","))
    min_z = min(b1[-1], b2[-1])
    max_x = max(b1[0], b2[0])
    max_y = min(b1[1], b2[1])

    return (b1, b2), max_x, max_y, min_z
```

```python
blocks = []
max_x_v = []
max_y_v = []
min_z_v = []
for line in lines: 
    block, max_x, max_y, min_z = parse_line(line)
    max_x_v.append(max_x)
    max_y_v.append(max_y)
    min_z_v.append(min_z)
    blocks.append(block)
```

```python
m_x = max(max_x_v)
m_y = max(max_x_v)
```

```python
max_height = np.zeros((m_x+1, m_y+1), dtype=int)
block_index = -np.ones((m_x+1, m_y+1), dtype=int)
```

```python
order_blocks_fall = np.argsort(np.array(min_z_v))
```

```python
block_supports = []
for i in range(len(blocks)): 
    block_supports.append([])
```

```python
def fall_block_vertical(x, y, z_height, index): 
    max_height[(x, y)] = max_height[(x, y)] + z_height
    if block_index[(x, y)] != -1: 
        block_supports[block_index[(x, y)]].append(index)
    block_index[(x, y)] = index
```

```python
def fall_block_horizontal(c1, c2, b, index): 
    
    grid_to_check = []
    if b[1] == 0: 
        for i in range(min(c1[0], c2[0]), max(c1[0], c2[0])+1): 
            grid_to_check.append((i, c1[1]))
    else: 
        for i in range(min(c1[1], c2[1]), max(c1[1], c2[1])+1): 
            grid_to_check.append((c1[0], i))

    mh = -1
    #print("blocks to check")
    #print(grid_to_check)
    for x in grid_to_check: 
        if max_height[x] > mh: 
            candidates = []
            mh = max_height[x]
            if block_index[x] != -1: 
                candidates.append(block_index[x])
        elif max_height[x] == mh: 
            if block_index[x] != -1: 
                candidates.append(block_index[x])
    

    for x in grid_to_check: 
        max_height[x] = mh + 1
        block_index[x] = index
    
    for cand in candidates: 
        block_supports[cand].append(index)
    

    
```

```python
def fall_block(block, index): 
    #print(block)
    c1 , c2 = block 
    b = sub_tuples(c1 , c2)
    #print(b)
    #print(block_index)
    #print(max_height)
    if b[0] == 0 and b[1] == 0: 
        assert(c1[0] == c2[0])
        assert(c1[1] == c2[1])
        fall_block_vertical(c1[0] , c1[1], abs(b[2])+1, index)
    else: 
        assert(c1[2] == c2[2])
        fall_block_horizontal(c1 , c2, b, index)
```

```python
for i in order_blocks_fall: 
    block = blocks[i]
    fall_block(block, i)
```
