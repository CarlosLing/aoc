```python
from collections import Counter

def count_chars(string):
    return dict(Counter(string))
```

```python
ordered_cards = ["A", "K", "Q", "J", "T", "9", "8", "7", "6", "5", "4", "3", "2"]
card_score = {}
for i in range(len(ordered_cards)):   
    card_score[ordered_cards[i]] = len(ordered_cards)-i 
```

```python

def get_hand_score(hand): 
    hand_keys = count_chars(hand)
    q = 13
    combo = 0
    scores = [0, 0, 0, 0, 0, 0]
    for key, value  in hand_keys.items(): 
        scores[value] += np.power(q, card_score[key])
        if value == 2: 
            combo += 1
        elif value == 3: 
            combo += 4
        elif value == 4: 
            combo += 8
        elif value == 5: 
            combo += 16
    return(scores, combo)
```
