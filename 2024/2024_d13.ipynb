```python
paragraphs = data.get_paragraphs()[:-1]
parsed_data = []

for paragraph in paragraphs:
    lines = paragraph.split('\n')
    buttons = []
    for line in lines[:-1]:
        numbers = list(map(int, re.findall(r'\d+', line)))
        buttons.append(tuple(numbers))
    final_price = tuple(map(int, re.findall(r'\d+', lines[-1])))
    parsed_data.append((buttons, final_price))

parsed_data
```

```python
def det(x1, y1, x2, y2):
    return x1*y2 - x2*y1

def solve(ax, ay, bx, by, finalx, finaly):
    det_0 = det(ax, bx, ay, by)

    a = round(det(finalx, bx, finaly, by)/det_0)
    b = round(det(ax, finalx, ay, finaly)/det_0)

    if ax*a+bx*b == finalx and ay*a+by*b == finaly:
        return a*3+b
    return 0
```

```python
p1 = 0
p2 = 0
for machine in parsed_data:
    [(ax, ay), (bx, by)], (prizex, prizey) = machine
    p1 += solve(ax, ay, bx, by, prizex, prizey)

    prizex += 10000000000000
    prizey += 10000000000000
    p2 += solve(ax, ay, bx, by, prizex, prizey)
print(p1)
print(p2)
```
