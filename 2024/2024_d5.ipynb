```python
data_paragraphs
rules = data_paragraphs[0].split("\n")
rules = [[int(y) for y in x.split("|")] for x in rules]
sequences = data_paragraphs[1].split("\n")
sequences = [[int(y) for y in x.split(",")] for x in sequences[:-1]]
```

```python
def check_element(element, remaining_sequence, rules):
    for rule in rules: 
        if element == rule[1]: 
            if rule[0] in remaining_sequence: 
                return False
    return True

def check_sequence(sequence, rules):
    for i in range(len(sequence)):
        if not check_element(sequence[i], sequence[i+1:], rules):
            return False
    return True
```

```python
valid_sequences = [sequence for sequence in sequences if check_sequence(sequence, rules)]
```

```python
middle_sum = sum(sequence[len(sequence) // 2] for sequence in valid_sequences if sequence)
middle_sum
```

```python
not_valid_sequences = [sequence for sequence in sequences if not check_sequence(sequence, rules)]
```

```python
def check_element_rule(element, remaining_sequence, rules):
    for rule in rules: 
        if element == rule[1]: 
            if rule[0] in remaining_sequence: 
                return (False, rule[0])
    return (True, None)

def reorder_sequence(sequence, rules):
    for i in range(len(sequence)):
        check = False
        while not check:
            check, change_element = check_element_rule(sequence[i], sequence[i+1:], rules)
            if not check:
                change_element_position = sequence.index(change_element)
                sequence[i], sequence[change_element_position] = sequence[change_element_position], sequence[i]

```

```python
for sequence in not_valid_sequences: 
    reorder_sequence(sequence, rules)
```
